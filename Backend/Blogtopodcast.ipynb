```python
# methods to genereate audio for the text, based on the speaker we choose the voice
def generate_audio(text, speaker, i, folder):
  #print(text)
  #print(speaker)
  audio_file = os.path.join(folder, f"{i}_{speaker}.wav")
  if speaker == 'Host':
    voice = 'female'
  else:
    voice = 'male'
  # generate the audio file and save it to audio_file path
  tts.tts_to_file(text=text,
                file_path=audio_file,
                speaker_wav=f"/content/{voice}.wav",
                language="en")


def extract_text_from_url(input):
  # Send an HTTP GET request to the URL
  response = requests.get(input)

  # Check if the request was successful (status code 200)
  if response.status_code == 200:
    # Parse the HTML content of the webpage using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find and extract all the text content
    text = soup.get_text()

    # Print or manipulate the extracted text as needed
    return(text)
  else:
    return(f"Failed to retrieve the webpage. Status code: {response.status_code}")

def is_valid_url(url):
    try:
        result = urlparse(url)
        return all([result.scheme, result.netloc])
    except ValueError:
        return False

def custom_sort_key(file_name):
  # Split the file name into parts based on '_'
  parts = file_name.split('_')
  if len(parts) >= 2:
    try:
      # Extract the numeric part and convert it to an integer
      return int(parts[0])
    except ValueError:
      pass
  # If the above logic fails, return a large value to move it to the end
  return float('inf')


def generate_podcast_text(input, time, folder_to_save):
    words_limit = int(int(time) * 60 * 1.5)

    if is_valid_url(input):
      print('It is a valid url')
      prompt = extract_text_from_url(input)
    else:
      prompt = input

    # Define a conversation with a system message and user and assistant messages
    conversation = [
        {"role": "system", "content": "You are a helpful assistant. who can create a conversation like podcast bewteen a Sarah who is host of the podcast and pick any random male name for the guest, podcast should be on a topic which is asked by the user, you can also use the information given by user to create conversation in the podcast. before giving response give the information that who is host and who is guest of the podcast."},
        {"role": "user", "content": f"Create a podcast on {input} with atmost {words_limit} words as limit"},
    ]

    # Generate a response from the assistant
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",  # Use the gpt-3.5-turbo model for faster results
        messages=conversation
    )

    # Extract and print the assistant's reply
    assistant_reply = response['choices'][0]['message']['content']
    # print(assistant_reply)


    # Write string content to a text file
    with open(f"{folder_to_save}/conversation.txt", "w") as file:
        file.write(assistant_reply)

    # Read the content from the text file
    # with open("conversation.txt", "r") as file:
    #    conversation = file.read()

    return(assistant_reply)

def generate_audio_from_conversation(conversation, folder_to_save):
    print('came to start generation of audio')
    texts = conversation.split('\n')
    texts = [each for each in texts if each != '']

    # Get the current timestamp
    # timestamp = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")

    # Define the folder name with the timestamp
    #folder = f"folder_{timestamp}"

    audio_folder = f"{folder_to_save}/audio"

    try:
        os.mkdir(audio_folder)
        print(f"{audio_folder} Folder created ")

    except FileExistsError:
        parse_schema



    for i in range(len(texts)):
        if ': ' in texts[i]:
            speaker, text = texts[i].split(': ',1)
            if 'Host' in speaker or 'Sarah' in speaker:
                speaker = 'Host'
            else:
                speaker = 'Guest'
            generate_audio(text, speaker, i, audio_folder)
            print(f"done_{i}")

    files = [each for each in os.listdir(audio_folder) if each.endswith('.wav')]
    files = sorted(files, key=custom_sort_key)

    output_audio = AudioSegment.empty()

    # Loop through the input files and append them with a 1-second gap
    for input_file in files:
        if 'Host' in input_file or 'Guest' in input_file:
            # Load the current input file
            audio = AudioSegment.from_wav(os.path.join(audio_folder,input_file))


            # Append the current audio and the gap to the output_audio
            output_audio += audio #+ silence_segment

    # Save the concatenated audio as a single WAV file
    output_audio.export(os.path.join(audio_folder,"final.wav"), format="wav")


def runcmd(cmd, verbose = False, *args, **kwargs):
    process = subprocess.Popen(
        cmd,
        stdout = subprocess.PIPE,
        stderr = subprocess.PIPE,
        text = True,
        shell = True
    )
    std_out, std_err = process.communicate()
    if verbose:
        print(std_out.strip(), std_err)
    pass


def generate_video(audio_path, video_path):
  runcmd(f"cd /content/Wav2Lip && python inference.py --checkpoint_path checkpoints/wav2lip_gan.pth --face {video_path} --audio {audio_path} --resize_factor 2", verbose=True)
  return('/content/Wav2Lip/results/result_voice.mp4')


def generate_video_from_audio(audio_files, audio_files_path, video_files_path):

  try:
      os.mkdir(video_files_path)
      print(f"created {video_files_path}")
  except FileExistsError:
      print(f"Folder already exists")

  for each in audio_files:

    audio_path = os.path.join(audio_files_path,each)

    print(f'running for file {audio_path}')

    if 'Guest' in each:
      video_path = '/content/male_guest_iv1.mp4'
    else:
      video_path = '/content/female_host_iv1.mp4'
    output_video = generate_video(audio_path, video_path)
    print(f"{audio_path.split('/')[-1].split('.')[0]}.mp4")
    shutil.copy(output_video, f"{video_files_path}/{audio_path.split('/')[-1].split('.')[0]}.mp4")


  video_files = [each for each in os.listdir(video_files_path) if 'Host' in each or 'Guest' in each]
  video_files = sorted(video_files, key=custom_sort_key)

  # Initialize a list to store VideoFileClip objects
  video_clips = []


  # Load and append each video file to the list, setting duration explicitly
  for _ in video_files:
      video_file = os.path.join(video_files_path, _)
      video_clip = VideoFileClip(video_file)
      video_clips.append(video_clip)

  # Concatenate the video clips with transitions
  final_clip = concatenate_videoclips(video_clips, method="compose")
  # If you want to keep the original audio from the videos
  audio_clip = AudioFileClip(f"{audio_files_path}/final.wav")

  final_clip = final_clip.set_audio(audio_clip)
  final_clip.write_videofile(f"{video_files_path}/final.mp4")
```

```python

thread_event = threading.Event()

def generate_podcast_text_bg(input_string, input_time, folder_to_save):
  if thread_event.is_set():
    generate_podcast_text(input_string, input_time, folder_to_save)
    thread_event.clear()

def generate_audio_bg(username, conversation, folder_to_save):
  if thread_event.is_set():
    generate_audio_from_conversation(conversation, folder_to_save)
    thread_event.clear()

def generate_video_bg(audio_files, audio_files_path, video_files_path):
  if thread_event.is_set():
    generate_video_from_audio(audio_files, audio_files_path, video_files_path)
    thread_event.clear()


# # Send the task
# result = generate_audio.apply_async(args=(username, podcast_text))
```

```python
app = Flask(__name__)
CORS(app)
run_with_ngrok(app)


@app.route("/podcast_text_generation/", methods=["POST"])
def podcast_text_generation():
    username = request.form["username"]
    input_string = request.form["input_string"]
    input_time = request.form["input_time"]

    user_folder = f"/content/{username}"
    try:
        os.mkdir(user_folder)
        print(f"created {user_folder}")
    except FileExistsError:
        print(f"Folder already exists")

    request_time_stamp = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")

    folder_to_save = f"{user_folder}/{request_time_stamp}"
    try:
        os.mkdir(folder_to_save)
        print(f"created {folder_to_save}")
    except FileExistsError:
        print(f"Folder already exists")

    # Send an acknowledgment to the user
    response = {"message": f"Podcast generation started for user: {username}",
                "response_unique_key": f"{input_time}"}
    # You can return this response immediately

    try:
      if not thread_event.is_set():  # Check if the thread is not already running
        thread_event.set()
        thread = threading.Thread(target=generate_podcast_text_bg, args=[input_string, input_time, folder_to_save])
        thread.start()
        return jsonify(response)
      else:
        return jsonify({"message":"Server is currently busy can you try again later"})
    except Exception as error:
      return str(error)



@app.route("/fetch_podcast_text_generated/", methods=["POST"])
def fetch_podcast_text_generated():
    username = request.form["username"]
    request_time_stamp = request.form["request_time_stamp"]

    user_folder = f"/content/{username}"

    folder_to_save = f"{user_folder}/{request_time_stamp}/conversation.txt"
    print(folder_to_save)
    try:
      if os.path.isfile(folder_to_save):
        # Read the content from the text file
        with open(f"{folder_to_save}", "r") as file:
            conversation = file.read()

        response = {"username": username, "podcast_text": conversation}
        return jsonify(response)
      else:
        return jsonify({"message":f"Podcast text is not yet generated, please wait for some time"})
    except Exception as error:
      return str(error)



@app.route("/audio_file_generation/", methods=["POST"])
def audio_file_generation():
    username = request.form["username"]
    podcast_text = request.form["podcast_text"]
    request_time_stamp = request.form["request_time_stamp"]


    user_folder = f"/content/{username}"

    folder_to_save = f"{user_folder}/{request_time_stamp}"


    # Read the content from the text file
    with open(f"{folder_to_save}/conversation.txt", "r") as file:
        conversation = file.read()

    # Send an acknowledgment to the user
    response = {"message": "Audio generation started for user: " + username}
    # You can return this response immediately

    try:
      if not thread_event.is_set():  # Check if the thread is not already running
        thread_event.set()
        thread = threading.Thread(target=generate_audio_bg, args=[username, conversation, folder_to_save])
        thread.start()
        return jsonify(response)
      else:
        return jsonify({"message":"Server is currently busy can you try again later"})
    except Exception as error:
      return str(error)



@app.route("/video_file_generation/", methods=["POST"])
def video_file_generation():
    username = request.form["username"]
    podcast_text = request.form["podcast_text"]
    request_time_stamp = request.form["request_time_stamp"]


    user_folder = f"/content/{username}"

    folder_to_save = f"{user_folder}/{request_time_stamp}"

    audio_files_path = f"/content/{username}/{request_time_stamp}/audio"
    video_files_path = f"/content/{username}/{request_time_stamp}/video"

    # Read the content from the text file


    if not os.path.isfile(f"{audio_files_path}/final.wav"):
      return jsonify({"message":"Audio is not yet generated"})

    audio_files = [each for each in os.listdir(audio_files_path) if 'Host' in each or 'Guest' in each]
    audio_files = sorted(audio_files, key=custom_sort_key)


    # Send an acknowledgment to the user
    response = {"message": "Video generation started for user: " + username}
    # You can return this response immediately

    try:
      if not thread_event.is_set():  # Check if the thread is not already running
        thread_event.set()
        thread = threading.Thread(target=generate_video_bg, args=[audio_files, audio_files_path, video_files_path])
        thread.start()
        return jsonify(response)
      else:
        return jsonify({"message":"Server is currently busy can you try again later"})
    except Exception as error:
      return str(error)

@app.route("/fetch_audio_file_generated/", methods=["POST"])
def fetch_audio_file_generated():
    username = request.form["username"]
    podcast_text = request.form["podcast_text"]
    request_time_stamp = request.form["request_time_stamp"]

    audio_file_path = f"/content/{username}/{request_time_stamp}/audio/final.wav"
    try:
      if os.path.isfile(audio_file_path):
        return send_file(
          audio_file_path,
          mimetype="audio/wav",
          as_attachment=True)
      else:
        return jsonify({"message":f"Audio file not yet generated please wait for a while"})
    except Exception as error:
      return jsonify({"message":f"Some error interupted the generation of audio"})



@app.route("/fetch_video_file_generated/", methods=["POST"])
def fetch_video_file_generated():
    username = request.form["username"]
    podcast_text = request.form["podcast_text"]
    request_time_stamp = request.form["request_time_stamp"]

    video_file_path = f"/content/{username}/{request_time_stamp}/video/final.mp4"
    try:
      if os.path.isfile(video_file_path):
        return send_file(
          video_file_path,
          mimetype="video/mp4",
          as_attachment=True)
      else:
        return jsonify({"message":f"Video file not yet generated please wait for a while"})
    except Exception as error:
      return jsonify({"message":f"Some error interupted the generation of Video"})



@app.route("/fetch_history/", methods=["POST"])
def fetch_history():
    username = request.form["username"]

    main_folder = f"/content/{username}"  # Replace with the path to your main folder

    dict = {}

    if os.path.exists(main_folder) and os.path.isdir(main_folder):
        folder_list = [f for f in os.listdir(main_folder) if os.path.isdir(os.path.join(main_folder, f))]

        for folder in folder_list:
            status = {}
            conversation_file = f"{main_folder}/{folder}/conversation.txt"
            audio_file = f"{main_folder}/{folder}/audio/final.wav"
            video_file = f"{main_folder}/{folder}/video/final.mp4"

            if os.path.isfile(conversation_file):
                status['Conversation'] = True
            else:
                status['Conversation'] = False

            if os.path.isfile(audio_file):
                status['Audio'] = True
            else:
                status['Audio'] = False

            if os.path.isfile(video_file):
                status['Video'] = True
            else:
                status['Video'] = False

            dict[folder] = status

        return jsonify(dict)


    else:
        return jsonify({"message":f"No history for the {username}"})



@app.route("/")
def read_root():
    return jsonify({"message": "Hello, world!"})

if __name__ == '__main__':
    app.run()
```
